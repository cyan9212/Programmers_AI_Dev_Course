# 자전거 대여 수요 예측

- `datetime`: 날자
- `season`: 1 = 봄, 2 = 여름, 3 = 가을, 4 = 겨울
- `holiday`: 1 = 주말을 제외한 공휴일, 0 = 휴일이 아닌 날
- `workingday`: 1 = 주말 및 휴일이 아닌 주중, 0 = 주말 및 휴일
- `weather`: 1 = 맑음/약간 흐림, 2 = 안개/안개+흐림, 3 = 가벼운 눈/가벼운 비+천둥
- `temp`: 온도(섭씨)
- `atemp`: 체감온도(섭씨)
- `humidity`: 상대습도
- `windspeed`: 풍속
- `casual`: 등록되지 않은 사용자가 대여한 횟수
- `registered`: 등록된 사용자가 대여한 횟수
- `count`: 대여 횟수

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
bike_df = pd.read_csv('/jupyter/Data/bike-sharing-demand/train.csv')
print('DataFrame shape: ',bike_df.shape)
bike_df.head()
```

```python
bike_df.info()
```

```python
# 문자열을 datetime 타입으로 변경
bike_df['datetime'] = bike_df.datetime.apply(pd.to_datetime)

# datetime 타입에서 년, 월, 일 시간을 추출
bike_df['year'] = bike_df.datetime.apply(lambda x: x.year)
bike_df['month'] = bike_df.datetime.apply(lambda x: x.month)
bike_df['day'] = bike_df.datetime.apply(lambda x: x.day)
bike_df['hour'] = bike_df.datetime.apply(lambda x: x.hour)
bike_df.head()
```

```python
# 불필요한 column을 drop
bike_df.drop(['datetime', 'casual', 'registered'], axis=1, inplace=True)
```

```python
bike_df.groupby(['season','count'])['count'].mean()
```
