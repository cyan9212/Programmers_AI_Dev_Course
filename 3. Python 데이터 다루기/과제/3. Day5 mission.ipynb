### Bonus Part
- 가장 많은 작품이 올라간 국가는 어디이고, 얼마나 많은 작품이 있는가?

```python
# country column의 결측치가 없는 데이터만 추출
new_df = df[df['country'].isnull()==False] 

# 각 국가별 작품의 수를 나타내는 딕셔너리 선언
dic = {}

# country column에 여러개의 국가를 가진 데이터를 처리
for data in new_df['country']:
    countries = data.split(',')
    for country in countries:
        country = str(country)
        if country and country[0] == ' ':
            country = country[1:]
        dic[country] = dic.get(country,0) +1 
        
print(dic)
```

United States에서 올라온 작품이 3690개로 가장 많다.

## Mission 2. 가즈아!

```python
coin_df = pd.read_csv('/jupyter/Data/Bitcoin/BitCoin.csv')
coin_df.head(20)
```

### Mandatory Part
- 2016.6 ~ 2017.6 기간의 5-MA(Moving Average) 비트코인 가격 그래프를 그려주세요

```python
from datetime import datetime

# 매달 1일~12일까지의 잘못된 날짜 데이터를 수정
past = '2017-09-23'
new_Date = []
for date in coin_df['Date']:
    year,month,day = date.split('-')
    a = datetime(int(year),int(month),int(day))
    p_year,p_month,p_day = past.split('-')
    b = datetime(int(p_year),int(p_month),int(p_day))
    if (b-a).days == 1:
        new_Date.append(date)
        past = date
        continue
    new_date = year+'-'+day+'-'+month
    new_Date.append(new_date)
    past = new_date
    
coin_df['new_Date'] = new_Date
coin_df[['Date', 'new_Date']][10:23]
```

```python
# 잘못된 날짜 데이터를 포함하는 Date column을 제거

coin_df.drop('Date', axis=1, inplace=True)
coin_df
```

```python
# new_Date column의 값이 2016-06-01 ~ 2017-06-30인 데이터들만 추출

new_coin_df = coin_df[(coin_df['new_Date']>='2016-06-01') & (coin_df['new_Date']<'2017-07-01')]
new_coin_df = new_coin_df.sort_values(by='new_Date')
new_coin_df.reset_index(drop=True)
```

```python
plt.figure(figsize=(10,5))
plt.plot(new_coin_df['new_Date'], new_coin_df['Open'].rolling(window=5).mean(), color='#f2a900', label='5-MA')

# 매월 1일을 xtick으로 설정

xtick = []
for date in new_coin_df['new_Date']:
    year,month,day = date.split('-')
    if day == '01':
        xtick.append(date)
        
plt.xticks(xtick,rotation=45)

plt.title('BitCoin Price', size=20)
plt.xlabel('Date', size=15)
plt.ylabel('Price($)', size=15)
plt.axis(['2016-06-01', '2017-06-30', 0, 3000])
plt.legend()
plt.show()
```
